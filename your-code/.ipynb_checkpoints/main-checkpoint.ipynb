## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform
x = uniform.rvs(size=10)
a = 2
b = 3
randoms  = a + (b-a)*x
print(randoms)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generates_distuni_rand(bottom, ceiling, count):
    my_uniform = uniform(bottom, ceiling)
    uniforme_count = my_uniform.rvs(size=count)
    return uniforme_count
```

```python
import seaborn as sns
```

```python
dist1 = generates_distuni_rand(10,15,100)
sns.histplot(dist1, color = "gray", bins = 10);
```

```python
dist2 = generates_distuni_rand(10,60,1000)
sns.histplot(dist2, color = "gray", bins = 10);
```

How are the two distributions different?

The second one has a more constant distribution due to the size of the sample is bigger.

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm
```

```python
def generates_distnom_rand(average, std_dev, count):
    my_normal = norm(loc=average, scale=std_dev)
    normal_count = my_normal.rvs(size=count)
    return normal_count
```

```python
distnom1 = generates_distnom_rand(10,1,1000)
sns.histplot(distnom1, kde = True, color = "grey");
```

```python
distnom2 = generates_distnom_rand(10,50,1000)
sns.histplot(distnom2, kde = True, color = "grey");
```

How are the two distributions different?

Second distribution has a widely behaviour due to it has a bigger standar deviation

3. Combined MPG

```python
sns.histplot(df["Combined MPG"], kde = True, color = "grey");
```

Which one(s) of the variables are nearly normally distributed? How do you know?

```python
from statsmodels.graphics.gofplots import qqplot
import matplotlib.pyplot as plt
```

Lets check it visually with the QQ graph for each distribution

```python
qqplot(df["Fuel Barrels/Year"] , line='s')
plt.show()
```

```python
qqplot(df["CO2 Emission Grams/Mile"] , line='s')
plt.show()
```

```python
qqplot(df["Combined MPG"] , line='s')
plt.show()
```

It seems that the "CO2 Emission Grams/Mile" could satisfy the normal distribution, the rest of them is more difficult to say they do as well. Let's check it with the K2 test!!

```python
from scipy.stats import normaltest
```

```python
distributions = ["Fuel Barrels/Year","CO2 Emission Grams/Mile","Combined MPG"]
```

```python
for dist in distributions:
    # Test de DAgostino
    stat, p = normaltest(df[dist])
    print(f'\n{dist} Distribution -> Estadistico=%.3f, p=%.3f' % (stat, p))
    # Interpretación
    alpha = 0.05
    if p > alpha:
       print(f'La muestra {dist} parece Gaussiana o Normal ')
    else:
       print(f'La muestra {dist} no parece Gaussiana o Normal')
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python

from scipy.stats import expon
```

```python
def generates_distexp_rand(count):
    mmy_e = expon(scale=10)
    expon_count = mmy_e.rvs(size=count)
    return expon_count

```

```python
disexp1 = generates_distexp_rand(10)
sns.histplot(disexp1, kde = True, color = "red");
```

```python
disexp2 = generates_distexp_rand(100)
sns.histplot(disexp2, kde = True, color = "red");
```

How are the two distributions different?

The first one is more flat due to it has less steps to create the exponential effect with the same average of occurrence
